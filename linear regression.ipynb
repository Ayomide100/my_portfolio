#### Predictive Modeling
* Split dataset into X and Y
* We scale input X
* We split X and Y into 70% by 30% for training and test
* Import algorithm and the evaluation metric method
* Train the algorithm on training set
* Evaluate model on test set

```python
# Rule 1
X = df.drop("Item_Outlet_Sales", axis = 1)
Y= df["Item_Outlet_Sales"]
```

```python
print("X shape is", X.shape)
print("Y shape is", Y.shape)
```

```python
# Rule 2
from sklearn.preprocessing import MinMaxScaler
# Initialize minmaxscaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
#Rule 3
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X, Y, random_state = 14)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
#Rule 4
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as MSE
```

```python
#Rule 5
#Initialize the model
clf = KNN(n_neighbors= 38)
#Fiting the model on the training set
clf.fit(train_x, train_y)
```

```python
clf.predict(test_x)
```

```python
predicted_output = clf.predict(test_x)
```

```python
# Rule 6
MSE(predicted_output, test_y)
```

```python
errorList = []
for i in range(1, 50):
    clf = KNN(n_neighbors= i)
    #Fiting the model on the training set
    clf.fit(train_x, train_y)
    predicted_output = clf.predict(test_x)
    error = MSE(predicted_output, test_y)
    myDict = {i: error}
    errorList.append(myDict)
```

```python
errorList
```

```python
errorList = []
k_values = []
for i in range(1, 51):
    clf = KNN(n_neighbors= i)
    #Fiting the model on the training set
    clf.fit(train_x, train_y)
    predicted_output = clf.predict(test_x)
    error = MSE(predicted_output, test_y)
    k_values.append(i)
    errorList.append(error)
```

```python
plt.plot(k_values, errorList, c = "red")
plt.xlabel("K values")
plt.ylabel("Errors")
plt.title("MSE on big mart sales using KNN algorithm", {'fontsize' : 10})
plt.show()
```
