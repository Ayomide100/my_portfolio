```python
import matplotlib.pyplot as plt
```

```python
myList = [1, 3, 5, 7, 10, 11]
```

### Line chart

```python
plt.plot(myList)
plt.ylabel("My List")
plt.xlabel("Index")
plt.show()
```

```python
myListTwo = [10, 20, 30, 40, 50, 60]
```

```python
plt.plot(myList, myListTwo)
plt.xlabel("First List")
plt.ylabel("Second List")
plt.title("List one VS List two")
plt.show()
```

```python
plt.plot(myList, myListTwo, "ro")
plt.xlabel("First List")
plt.ylabel("Second List")
plt.title("List one VS List two")
plt.show()
```

```python
import numpy as np
```

```python
myArrOne = np.array(myList)
myArrTwo = np.array(myListTwo)
myArrOne
```

```python
plt.plot(myArrOne, myArrTwo, myArrOne, myArrOne ** 2)
plt.show()
```

### Scatter plot

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data
```

```python
data['a']
```

```python
plt.scatter(data['a'], data['c'], c =data['d'], s= data['a'])
plt.xlabel("A")
plt.ylabel("C")
plt.title("Scatter plot")
plt.show()
```

```python
x = ['a', 'b', 'c']
y = [2, 5, 10]
```

```python

plt.figure(figsize= (9, 3))
plt.subplot(1,3,1)
plt.bar(x, y, color = "orange")
# plt.gcf().set_facecolor('#000000') 
plt.title("fig one")
plt.subplot(1,3,2)
plt.scatter(x, y)
plt.title("fig two")
plt.subplot(1,3,3)
plt.plot(x, y, c = "red")
plt.title("fig three")
plt.show()
```

```python
plt.figure(figsize= (3, 9))
plt.subplot(3,1,1)
plt.bar(x, y)
plt.subplot(3,1,2)
plt.scatter(x, y)
plt.subplot(3,1,3)
plt.plot(x, y)
plt.show()
```
