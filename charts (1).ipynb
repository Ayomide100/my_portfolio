### Scatter plot

```python
height = [10, 40, 20, 67, 75, 11]
weight = [75.7, 80.2, 62.5, 73, 68.2, 100]
```

```python
plt.scatter(height, weight, c= '#000', s = height)
plt.show()
```

### Bar chart

```python
sex = ["boy", "Girl"]
Age = [12, 24]
```

```python
plt.bar(sex, Age)
plt.show()
```

```python
plt.barh(sex, Age)
plt.ylabel("Sex")
plt.xlabel("Age")
plt.show()
```

```python
# Plotting a 2 by 2 dimensional figure (It will contain four charts)
plt.figure(figsize= (15, 5))
plt.subplot(2,2,1)
plt.plot(x, y, '-ro', b, a, 'g^', x , y**2, 'bs' )
plt.subplot(2,2,2)
plt.scatter(height, weight, c= '#000', s = height)
plt.subplot(2,2,3)
plt.bar(sex, Age)
plt.subplot(2,2,4)
plt.barh(sex, Age)
plt.ylabel("Sex")
plt.xlabel("Age")
plt.suptitle("A two by two dimensional plot")
plt.show()
```
