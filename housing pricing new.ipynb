```python
def converter(price):
    if 'cr' in price.lower(): 
        a = float(price.lower().replace('cr', '')) * 10000000
        return a
    elif 'lac' in price.lower():
        b = float(price.lower().replace('lac', '')) * 100000
        return b
    else:
        return None
```

### Convert cover_area values into squareft

```python
df['Carpet Area'].unique()
```

```python
def convert_sqft(area):
    if 'sqft' in area:
        a = float(area.lower().replace('sqft', ''))
        return a
    elif 'sqyrd' in area:
        b = float(area.lower().replace('sqyrd', '')) * 9
        return b
    else:
        return None
```

```python
convert_sqft('500.2')
```

```python
# Convert values under carpet area to string as apply rhe convert_sqft function
df['Carpet Area'] = df['Carpet Area'].astype('str').apply(convert_sqft)
```
